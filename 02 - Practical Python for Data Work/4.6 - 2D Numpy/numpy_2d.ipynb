## Tujuan
- Membuat dan mengakses array NumPy 2 dimensi (matriks)
- Operasi dasar matriks: addition, scalar multiplication, Hadamard product, dot product
- Transpose matriks

```python
import numpy as np

a = [[11, 12, 13], [21, 22, 23], [31, 32, 33]]
a
```

```python
A = np.array(a)
A
```

```python
print("ndim:", A.ndim)   # jumlah dimensi/axes
print("shape:", A.shape) # (baris, kolom)
print("size:", A.size)    # total elemen
```

```python
# baris index 1, kolom index 2
print(A[1, 2])
print(A[1][2])   # cara alternatif, hasil sama
```

```python
A[0][0]
```

```python
A[0][0:2]
```

```python
A[0:2, 2]
```

## Basic Operations

```python
X = np.array([[1, 0], [0, 1]])
Y = np.array([[2, 1], [1, 2]])

Z = X + Y
Z
```

```python
Y = np.array([[2, 1], [1, 2]])
Z = 2 * Y
Z
```

```python
X = np.array([[1, 0], [0, 1]])
Y = np.array([[2, 1], [1, 2]])

Z = X * Y   # perkalian per posisi, BUKAN perkalian matriks matematis
Z
```

```python
A = np.array([[0, 1, 1], [1, 0, 1]])
B = np.array([[1, 1], [1, 1], [-1, 1]])

Z = np.dot(A, B)   # A shape (2,3), B shape (3,2) -> hasil shape (2,2)
Z
```

```python
np.sin(Z)
```

```python
C = np.array([[1, 1], [2, 2], [3, 3]])
C
```

```python
C.T   # tukar posisi baris jadi kolom, dan sebaliknya
```
