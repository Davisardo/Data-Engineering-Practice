# 5.6 - Practice Project GDP Data

## Practice Project: GDP Data Extraction and Processing

## Skenario
Sebagai junior Data Engineer, buat script untuk mengekstrak 10 negara dengan GDP terbesar (dalam Miliar USD, dibulatkan 2 desimal) menurut IMF.

## Tujuan
- Pakai Pandas untuk baca data dari halaman HTML sebagai DataFrame
- Pakai NumPy untuk manipulasi data
- Load DataFrame hasil ke file CSV

```python
import numpy as np
import pandas as pd

# suppress warning
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')
```

```python
URL = "https://web.archive.org/web/20230902185326/https://en.wikipedia.org/wiki/List_of_countries_by_GDP_%28nominal%29"
```

### Exercise 1
Ekstrak data GDP dari URL menggunakan Web Scraping.

```python
# Ekstrak semua tabel dari halaman web, tabel yang dibutuhkan adalah tabel ke-3 (index 3)
tables = pd.read_html(URL)
df = tables[3]

# Ganti nama kolom jadi angka (karena header aslinya multi-level, bikin bingung)
df.columns = range(df.shape[1])

# Ambil kolom index 0 (nama negara) dan 2 (GDP menurut IMF)
df = df[[0, 2]]

# Ambil baris index 1-10 (10 negara teratas, baris 0 biasanya "World")
df = df.iloc[1:11, :]

# Beri nama kolom yang jelas
df.columns = ['Country', 'GDP (Million USD)']

df
```

### Exercise 2
Ubah kolom GDP dari Juta USD ke Miliar USD, dibulatkan 2 desimal. Ubah nama header jadi "GDP (Billion USD)".

```python
# Pastikan tipe data GDP adalah integer
df['GDP (Million USD)'] = df['GDP (Million USD)'].astype(int)

# Konversi dari Juta ke Miliar (bagi 1000)
df[['GDP (Million USD)']] = df[['GDP (Million USD)']] / 1000

# Bulatkan 2 desimal
df[['GDP (Million USD)']] = np.round(df[['GDP (Million USD)']], 2)

# Ganti nama kolom
df = df.rename(columns={'GDP (Million USD)': 'GDP (Billion USD)'})

df
```

### Exercise 3
Simpan DataFrame ke file CSV bernama "Largest_economies.csv".

```python
df.to_csv('./Largest_economies.csv')
print("Data berhasil disimpan ke Largest_economies.csv")
```
