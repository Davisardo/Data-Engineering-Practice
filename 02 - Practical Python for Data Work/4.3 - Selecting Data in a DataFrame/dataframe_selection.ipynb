## Tujuan
- Memahami struktur DataFrame dan Series di Pandas
- Memilih kolom, baris, dan cell tertentu menggunakan `loc()` dan `iloc()`
- Melakukan slicing pada DataFrame

```python
import pandas as pd
```

```python
# Dictionary sebagai data mentah
x = {
    'Name': ['Rose', 'John', 'Jane', 'Mary'],
    'ID': [1, 2, 3, 4],
    'Department': ['Architect Group', 'Software Group', 'Design Team', 'Infrastructure'],
    'Salary': [100000, 80000, 50000, 60000]
}

# ubah dictionary jadi DataFrame -> keys jadi nama kolom, values jadi isi baris
df = pd.DataFrame(x)
df
```

```python
# Ambil kolom "ID" -> pakai DOUBLE bracket [[...]] supaya hasilnya tetap DataFrame
x = df[['ID']]
x
```

```python
type(x)
```

```python
# Ambil 3 kolom: Department, Salary, ID (urutan bebas, tidak harus sama seperti df asli)
z = df[['Department', 'Salary', 'ID']]
z
```

## Exercise 2: loc() dan iloc()

- `iloc[row, col]` — akses berdasarkan **posisi angka**
- `loc[row, col]` — akses berdasarkan **nama/label**

```python
# Baris pertama (index 0), kolom pertama (index 0)
df.iloc[0, 0]
```

```python
# Baris pertama, kolom ketiga (index 2)
df.iloc[0, 2]
```

```python
# Baris dengan index 0, kolom bernama 'Salary'
df.loc[0, 'Salary']
```

```python
# Buat DataFrame baru df2, set kolom "Name" jadi index
df2 = df
df2 = df2.set_index("Name")
```

```python
df2.head()
```

```python
# Sekarang index-nya nama orang, bukan angka
df2.loc['Jane', 'Salary']
```

## Exercise 3: Slicing

Perhatikan: `iloc` slicing TIDAK termasuk batas akhir, tapi `loc` slicing TERMASUK batas akhir.

```python
# baris index 0-1 (2 tidak termasuk), kolom index 0-2 (3 tidak termasuk)
df.iloc[0:2, 0:3]
```

```python
# baris index 0 SAMPAI 2 (termasuk 2!), kolom 'ID' sampai 'Department' (termasuk)
df.loc[0:2, 'ID':'Department']
```

```python
# baris 'Rose' sampai 'Jane' (termasuk Jane), kolom 'ID' sampai 'Department'
df2.loc['Rose':'Jane', 'ID':'Department']
```
